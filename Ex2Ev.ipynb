Complete la siguiente celda para poder otbener los estadísticos necesarios para poder evaluar el modelo y conteste a las preguntas despues de los estadísticos para realizar una conclusión final

```python
def plot_confusion_matrix(test_labels, target_predicted):
    matrix = confusion_matrix(test_labels, target_predicted)
    df_confusion = pd.DataFrame(matrix)
    colormap = sns.color_palette("BrBG", 10)
    sns.heatmap(df_confusion, annot=True, fmt='.2f', cbar=None, cmap=colormap)
    plt.title("Confusion Matrix")
    plt.tight_layout()
    plt.ylabel("True Class")
    plt.xlabel("Predicted Class")
    plt.show()

def plot_stats(test_labels, target_predicted):
    TN, FP, FN, TP = confusion_matrix(test_labels, target_predicted).ravel()

    sens  = float(TP)/(TP+FN)*100 # <INSERT CODE HERE>
    espc  = float(TN)/(TN+FP)*100# <INSERT CODE HERE>
    Vpp = float(TP)/(TP+FP)*100# <INSERT CODE HERE>
    Vpn = float(TN)/(TN+FN)*100# <INSERT CODE HERE>
    ACC = float(TP+TN)/(TP+FP+FN+TN)*100# <INSERT CODE HERE>
    classification_report_lr = classification_report(test_labels, target_predicted)

    print("Logistic Regression Metrics:")
    print("  Sensibilidad: ", sens, "%") 
    print("  Especificidad: ",espc, "%") 
    print("  Valor predictivo positivo ",Vpp, "%") 
    print("  Valor predictivo negativo: ",Vpn, "%") 
    print("  Eficacia: ",ACC, "%") 
    print('  Reporte del clasificador:')
    print(classification_report_lr)



plot_confusion_matrix(y_test, y_pred_lr)
plot_stats(y_test, y_pred_lr)
```
